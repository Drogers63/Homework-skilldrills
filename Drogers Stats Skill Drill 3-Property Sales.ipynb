```python
# Dependencies
import pandas as pd
import os
import scipy.stats as stats
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# created dataframe using csv fil
property_sales_df = pd.read_csv('/Users/debo/Downloads/property_sales.csv')
property_sales_df[['Downpayment','Gross_Sale']]
```

```python
# Create a boxplot to compare means
px.box(property_sales_df,x="Downpayment", y="Gross_Sale")
```

```python
# Extract individual groups
group1 = property_sales_df[property_sales_df["Downpayment"] == 1]["Gross_Sale"]
group2 = property_sales_df[property_sales_df["Downpayment"] == 1]["Gross_Sale"]
group3 = property_sales_df[property_sales_df["Downpayment"] == 1]["Gross_Sale"]
group4 = property_sales_df[property_sales_df["Downpayment"] == 1]["Gross_Sale"]
group5 = property_sales_df[property_sales_df["Downpayment"] == 1]["Gross_Sale"]
```

```python
# Perform the ANOVA
stats.f_oneway(group1, group2, group3, group4, group5)
```

```python
# the statistic value proves the Null Hypothesis is not correct. 
#A P value of of 1 shows weak evidence against the null hypothesis 
#so it ca't be rejected. The large P value is also an indicator
```
