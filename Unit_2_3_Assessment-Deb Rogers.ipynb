**Question 7: Generate a bar plot that visualizes the following Netflix show IMDB ratings: (5 points)**

`{"Series":["The Witcher","Disenchantment","Tiger King","Ozark","Stranger Things"],"User Ratings":[8.2,7.2,7.6,8.4,8.7]}`

````python
# Answer 7
import matplotlib.pyplot as plt
import numpy as np

#creating dataset
data ={'The Witcher': 8.2,'Disenchantment':7.2,'Tiger King':7.6,'Ozark':8.4,'Stranger Things':8.7}
series = list(data.keys())
ratings = list(data.values())

#bar plot
fig = plt.figure(figsize =(10,5))
plt.bar(series,ratings, color = 'blue',
       width = 0.5)         

  
plt.xlabel("Series") 
plt.ylabel("User Ratings") 
plt.title("Netflix Show IMDB Ratings") 
plt.show()
````
