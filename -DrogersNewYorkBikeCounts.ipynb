```python
# Dependencies
import pandas as pd
import os
from sklearn import datasets
import statsmodels.api as sm
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head()
```

```python
# DF showing two columns comparing temps to bikes crossing brooklyn bridge
bikecounts_df['Precipitation'] = bikecounts_df.Precipitation.astype(str)
```

```python
# DF showinf two columns comparing temps to bikes crossing brooklyn bridge
bikecounts_df[["High Temp (°F)","Precipitation", "Brooklyn Bridge"]].head()
```

```python
 # Select your independent X terms, and your dependent y term. 
X = bikecounts_df[['Brooklyn Bridge']]
y = bikecounts_df['High Temp (°F)']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
                   
#  sm.OLS(y.astype(float), X.astype(float)).fit()
```

```python
# Print the linear regression results
print(results.summary())
```

```python
# Print out the equation of a line
coef = results.params
print(f"The equation of our line is y = (Brooklyn Bridge *{coef[1]}) + {coef[0]}")
```
